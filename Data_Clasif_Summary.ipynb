# 3. Advanced Prompting Test (Few-Shot)

This stage focuses on testing whether providing context or few-shot examples improves the consistency and quality of Granite's sentiment classification.  
A few labeled examples are included within the prompt so the model can learn from the given structure and style before classifying new inputs.

The goal is to observe whether the model produces more precise labels and explanations when given prior examples of how sentiment should be categorized.

### Steps
1. Prepare a few labeled examples (positive, negative, and mixed) to serve as context.
2. Add these examples at the beginning of the prompt before listing the new reviews.
3. Run the same `invoke()` call using this enriched prompt.
4. Compare the new output with the previous “default” version.

### Expected Output
A clearer and more structured classification, potentially with improved reasoning consistency.  
The comparison will be stored in a simple table or Markdown summary highlighting the difference between the default and contextual outputs.

# 4. PARAMETER SET

| Parameter               | Description                                                                 | Default | Effect / Notes                                                                                  | Analogy / Example                                                                 |
|-------------------------|-----------------------------------------------------------------------------|---------|-------------------------------------------------------------------------------------------------|----------------------------------------------------------------------------------|
| **top_k**              | Limits candidate tokens considered for the next word.                       | 0       | Low (`30`) → focused and deterministic. High (`0`) → varied but less stable.                   | `0`: listens to everyone in a forum. `30`: listens to 30 most relevant people.  |
| **top_p**              | Controls diversity using nucleus sampling (select top tokens until probability reaches `p`). | 1.0     | Low (`0.7`) → directed and coherent. High (`1.0`) → creative and varied, sometimes less focused. | `1.0`: reads all opinions on Twitter. `0.7`: reads only top-liked tweets.        |
| **repetition_penalty**  | Penalizes repeated words or phrases.                                        | 1.0     | >1 → avoids repetition. Too high → may sound rigid.                                            | Prevents: "The phone is good. The phone is good. The phone is good."             |
| **max_tokens**         | Maximum number of tokens in output.                                         | -       | High → longer, detailed. Low → shorter, focused.                                             | Limits overly long responses.                                                    |
| **min_tokens**         | Minimum number of tokens in output.                                         | -       | Ensures output is informative enough.                                                         | -                                                                                |
| **random_seed**        | Controls randomness for consistent outputs.                                 | -       | Ensures output is the same across runs.                                                      | `42` → output stays consistent every run.                                        |
| **stopping_sequence**  | Stops text generation when a specific sequence is encountered.             | -       | Controls when the model stops writing.                                                       | `"\n\n"` → stops at two empty lines.                                             |
